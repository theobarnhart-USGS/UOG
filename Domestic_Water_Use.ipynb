```python
dat = pd.read_excel('./data/domestic_water_use_compilation_v1.xlsx')
```

```python
refs = dat.reference.unique()
refkeys = np.arange(1,len(refs)+1)

def set_refkey(ref):
    for r,refkey in zip(refs,refkeys):
        if r==ref:
            return refkey
```

```python
dat['refkey'] = dat.reference.map(set_refkey)
```

```python
dat.plot.scatter('refkey','value',figsize=(15,5))
plt.xticks(refkeys,refs,rotation=-25,ha='left',fontsize=14);
plt.xlabel('')
plt.ylabel('gallons/person/day', fontsize=14)
plt.title('Domestic Water Use', fontsize=18)
```

```python
dat.sort_values('year',ascending=True, inplace=True)
```

```python
vals = list(dat.value)
```

```python
refs = list(dat.reference)
```

```python
years = np.arange(1985,2030,5)
```

```python
labels = []

for year,ref in zip(years,refs):
    if year <= 2010:
        labels.append('%s\n%s*'%(year,ref))
    else:
        labels.append('%s'%(ref))
```

```python
plt.figure(figsize=(15,5))
plt.plot(years,vals,marker='o',markersize=10, ls='none')
#dat.plot('refkey','value',figsize=(15,5),legend=False, marker='o',markersize=10, ls='none')
plt.xticks(years,labels,rotation=-25,ha='left',fontsize=16);
plt.xlabel('')
plt.ylabel('gallons/person/day', fontsize=16)
plt.title('Domestic Water Use', fontsize=18)
plt.xlim(1982,2027)
plt.ylim(0,220)
plt.text(1983,20,'*USGS Water Use Compilation', fontsize=16)
```
